## Build Data Generator

```python
################################################################################
# Data Generator

# Defining generator functions for train/test samples
def get_input(path):     
    img = cv2.imread(path, )
    # Rotated1_image = imutils.rotate(img, angle=90)
    Rotated1_image = cv2.flip(img, 0)
    Rotated1_image = imutils.rotate(Rotated1_image, angle=-90)
    # Rotated1_image = img
    return Rotated1_image

def get_output(path):
    mask_1 = np.load(path)
    return mask_1["arr_0"]


def gen_pairs_train():
    data_paths = pd.read_csv('paths_lst.csv')
    for i in range(train_mount):
        # Get a random image each time
        idx = np.random.randint(0,train_mount)
        
        yield (get_input(data_paths["mixed_img"][idx]), get_output(data_paths["inf_mask_paths"][idx]))


def gen_pairs_test():
    for i in range(train_mount, val_mount):
        # Get a random image each time
        idx = np.random.randint(train_mount, val_mount)

        # x = tf.convert_to_tensor(get_input(data_paths["mixed_img"][idx]), dtype=tf.float32)
        # y = tf.convert_to_tensor(get_output(data_paths["inf_mask_paths"][idx]), dtype=tf.float32)
        x = get_input(data_paths["mixed_img"][idx])
        y = get_output(data_paths["inf_mask_paths"][idx])
        yield x, y

# Function to test input pipeline
sample_image, sample_label = next(gen_pairs_train())
```

```python
print(sample_image.shape, sample_label.shape)
```

```python
batch_size = 4
# Prepare the training dataset.
train_dataset = tf.data.Dataset.from_generator(generator=gen_pairs_train, output_types=(np.float32, np.float32))
train_dataset = train_dataset.batch(batch_size)

# Prepare the validation dataset.
test_dataset = tf.data.Dataset.from_generator(generator=gen_pairs_test, output_types=(np.float32, np.float32))
test_dataset = test_dataset.batch(batch_size)

```

## Training SSA_Net Model

```python
model = SSA_Net()
model.summary_model()
```

```python
# model.fit_generator(traingen, epochs=3)
# Define the optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
NUM_EPOCHS = 1
EPOCH_PER_DISPLAY = 1
total_loss = []
for epoch in range(NUM_EPOCHS):

    running_loss = []
    running_DS = []
    running_NSD = []
    running_HD = []
    running_MAE = []


    # Training
    for input, target in train_dataset:

        # Calculate and track graduents
        with tf.GradientTape() as tape:

            # Calculate model output and loss
            output = model(input, training=True)
            loss_ = evaluation_metrics.custom_loss(target, output)
            DS_ = evaluation_metrics.dice_Similarity_coef(target, output[0])
            NSD = normalized_surface_dice(target.numpy(), output[0].numpy(), 3)

            HD = evaluation_metrics.hausdoff_distance_func(target, output[0])
            MAE = evaluation_metrics.mean_absolute_error(target, output[0])

            # Tape gradients
            grads = tape.gradient(loss_, model.trainable_variables)
            
        # Track batch loss and accuracy
        running_loss.append(loss_)
        running_DS.append(DS_)
        running_NSD.append(NSD)
        running_HD.append(HD)
        running_MAE.append(MAE)


        # Optimize model based on the gradients
        optimizer.apply_gradients(zip(grads, model.trainable_variables))

    # Epoch calculations
    epoch_loss = np.mean(running_loss)
    epoch_DS = np.mean(running_DS)
    epoch_NSD = np.mean(running_NSD)
    epoch_HD = np.mean(running_HD)
    epoch_MAE = np.mean(running_MAE)


    if (epoch + 1) % EPOCH_PER_DISPLAY == 0:
        print(f"Epoch {epoch+1}: Loss: {epoch_loss:.4f} // Dice Coef: {epoch_DS:.4f} // NSD: {epoch_NSD:.4f} // HD: {epoch_HD:.4f} // MAE: {epoch_MAE:.4f}")
```

```python
def trust_module(y_pred: np.ndarray) -> np.float16:
    """
      Threshold our predicted array

      Arguments:
          y_pred -- Predicted output
      Returns:
          y_pred --  array with its class
    """

    for c in range(1, 4):
        y_p = y_pred[:, :, :, c-1]
        y_p[y_pred >= 0.95] = 1
    y_pred[y_pred < 1] = 0

    return y_pred
```
